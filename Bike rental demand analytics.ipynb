```python
#Transform datetime into listgio:
ym = []
for it in list(bikedf.loc[:,'datetime']):
    ym.append(it[0:7])
bikedf['yyyymm']=ym
```

```python
#Transform datetime into listgio:
listgio = []
for item in list(bikedf.loc[:,'datetime']):
    listgio.append(item[11:13])
bikedf['Gio']=listgio
```

# Correlation

```python
plt.figure(figsize=(15,10))
sns.heatmap(bikedf.corr(), annot = True)
#temp, atemp x casual, registered, count
```

# Trending over month

```python
# Bike rental demand growing trend. Inscrease over month and year.
plt.figure(figsize=(20,6))
sns.barplot(x="yyyymm", y="count",data=bikedf, palette='Blues')
```

```python
# Count in working day
plt.figure(figsize=(10,7))
sns.lineplot(x="Gio", y="count",hue='workingday',data=bikedf, palette=['r','b'])
#People rent Bike in Morning and Afternoon on Working day, maybe for work. Rent whole day in non-working day.

# Casual in working day
plt.figure(figsize=(10,8))
sns.lineplot(x="Gio", y="casual",hue='workingday',data=bikedf, palette=['r','b'])
# => Casual people didn't rent bike during working day. And higher in non-working day.
```

```python
# Hour x bike rental demand (breaking down by HOLIDAY): 
plt.figure(figsize=(10,6))
sns.lineplot(x="Gio", y="count",hue='holiday',data=bikedf, palette=['b','r'])

# People rent Bike at 7 (early morning), and NOON, and EVENING (highest demand) during holiday.
# in normal day, they rent in Morning and Evening (as same as Working day)

# Count in working day
plt.figure(figsize=(10,7))
sns.lineplot(x="Gio", y="count",hue='workingday',data=bikedf, palette=['r','b'])
#People rent Bike in Morning and Afternoon on Working day, maybe for work. Rent whole day in non-working day.
```

```python
#WEATHER slightly correlate with HUMIDITY 
plt.figure(figsize=(10,6))
sns.barplot(x="weather", y="humidity",data=bikedf, palette='rocket')

# 1: Clear, Few clouds, Partly cloudy, Partly cloudy 
# 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist    
# 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds 
# 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog 

# for humidity
plt.figure(figsize=(12,5))
sns.lineplot(x="humidity",y="count", data=bikedf, palette='rocket')
# People rental demand lower when the humidity is high (ofcourse it's related with weather is bad)

#People rent bike highest in Clear Weather and lowest when it's Light Snow, Light Rain
plt.figure(figsize=(10,6))
sns.barplot(x="weather", y="count",data=bikedf, palette='rocket')

#Fact: in the heavy rain/thunder storm. People rent bike higher than Light snow. 
# Why? => Might be they need to run out :D
```
